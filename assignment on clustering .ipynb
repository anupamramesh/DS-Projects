```python
# perform clustering by hierarchical method for crime data set
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns 
```

```python
crime=pd.read_csv("F:\\excelR\\assignments\\clustring\\crime_data.csv")
```

```python
crime
```

```python
# normalizing the data
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(crime.iloc[:,1:])
```

```python
df_norm
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='average'))
```

```python
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'average')
```

```python
hc
```

```python
y_hc = hc.fit_predict(df_norm)
y_hc
```

```python
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
crime['clust']=Clusters
```

```python
crime.head()
```

```python
crime.iloc[:,1:].groupby(crime.clust).mean()
```

### We can now clearly see the 4 number of clusters formed, which can be described as under

    1. Cluster1 = "0" = Defines that their are less murder rate and rape rate compared to other 3 clusters.
    2. Cluster2 = "1" = Defines that the rate of crime is 2nd highest among all the clsuters.
    3. Cluster3 = "2" = Defines that the rate of crime is highest among all the clsuters with less urban population.
    4. Cluster4 = "3" = Defines that the rate of crime is 3rd highest among all the clsuters.

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```python
df=pd.read_csv("F:\\excelR\\assignments\\clustring\\crime_data.csv")
```

```python
df
```

```python
df.info()
```

```python
array=df.values
```

```python
array
```

```python
stscaler=StandardScaler().fit(array)
X=stscaler.transform(array)
```

```python
X
```

```python
dbscan=DBSCAN(eps=0.9, min_samples=6)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```
